1. **What are lists in Python?**

  Lists in Python are ordered collections of items, where each item can be of any data type. They are one of the most versatile and commonly used data structures in Python. Lists are defined using square brackets [], and their elements are separated by commas.

2. Explain the working of lists:

  Lists in Python maintain the order of elements, meaning the position of an element in the list is preserved. Elements in a list can be accessed using indexing, and you can perform various operations on lists such as appending, inserting, updating, and deleting elements.

3. **What are List Containers?**

  List containers refer to data structures that hold a collection of items in an ordered sequence. Lists in Python are one of the primary examples of list containers. They provide methods and functionality to manipulate the stored items easily.

4. **What is dynamic insertion in the context of lists?**

  Dynamic insertion in the context of lists refers to the ability to add or insert items into a list without requiring predefined memory allocation. In Python, lists are dynamic in nature, which means you can add elements at any position without worrying about memory management.

5. **What are the advantages and disadvantages of lists?**

  ***Advantages of lists:***

  Dynamic size: Lists can grow or shrink as needed.
  Versatile: Lists can store heterogeneous data types.
  Order preservation: Elements are stored in the order they are added.
  Built-in methods: Lists come with various built-in methods for manipulation.

  **Disadvantages of lists:**

    Slower than arrays: Access and manipulation can be slower compared to arrays.
    Higher memory usage: Lists consume more memory due to additional metadata.
    Limited performance for large datasets: Lists may not be the most efficient for very large datasets.
  
5. **How is storage space allocated to lists? How dynamic size works?**

  Lists in Python are implemented as dynamic arrays. Initially, a small amount of memory is allocated to hold elements. As you append more items, if the list grows beyond its current allocated size, a new larger memory block is allocated, and elements are copied to the new block. This process allows lists to dynamically resize and accommodate more elements as needed. The growth factor for dynamic resizing can vary based on Python's implementation details.

## Slicing

Slicing is a technique used to extract a portion of a sequence, such as a list, string, or tuple, by specifying a range of indices. It allows you to create a new sequence containing the selected elements. Slicing is done using square brackets and the colon : operator.

Here are examples of slicing for lists, strings, and tuples:

Slicing for Lists:

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Extract elements from index 2 (inclusive) to index 5 (exclusive)
sliced_list = my_list[2:5]
print(sliced_list)  # Output: [3, 4, 5]
```

Slicing Strings

```python
my_string = "Hello, World!"

# Extract characters from index 7 (inclusive) to index 12 (exclusive)
sliced_string = my_string[7:12]
print(sliced_string)  # Output: "World"
```

Slicing for Tuples:


```python
my_tuple = (10, 20, 30, 40, 50)

# Extract elements from index 1 (inclusive) to index 4 (exclusive)
sliced_tuple = my_tuple[1:4]
print(sliced_tuple)  # Output: (20, 30, 40)
```

## Comprehensions

Comprehensions in Python are used to create new sequences, such as lists, dictionaries, and sets, in a concise and expressive way. They provide a more readable and efficient approach to generate sequences compared to using traditional loops and appending or adding elements manually. Comprehensions offer several benefits:

Conciseness: Comprehensions allow you to express complex operations in a single line of code, making your code more compact and easier to understand.

Readability: Comprehensions use a declarative style that closely resembles the mathematical notation for sets and lists. This makes the code more intuitive and readable, especially for simple operations.

Reduced Looping: Comprehensions encapsulate the loop logic within a compact syntax, eliminating the need for separate loop constructs. This leads to cleaner code and reduces the chances of introducing errors in loop iterations.

Performance: Comprehensions are often more efficient than traditional loops because they are implemented internally using optimized mechanisms. This can result in faster execution and better performance for large sequences.

Inline Transformation: Comprehensions allow you to apply transformations or filters to each element of a sequence while generating the new sequence. This makes it easy to perform operations like mapping, filtering, and applying conditions in a single step.

Overall, comprehensions enhance code readability, reduce the lines of code required, and improve the maintainability and efficiency of your programs, making them a preferred choice for creating sequences in Python.

## 1. Seperate Even & Odd

```python
def separate_even_odd(numbers):
    even_numbers = []  # List to store even numbers
    odd_numbers = []   # List to store odd numbers

    # Iterate through each number in the input list
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            even_numbers.append(num)  # Add even number to the even_numbers list
        else:
            odd_numbers.append(num)   # Add odd number to the odd_numbers list

    return even_numbers, odd_numbers  # Return the two separate lists

# Example list of numbers
numbers_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Separate even and odd numbers using the function
even_numbers, odd_numbers = separate_even_odd(numbers_list)

# Print the separate lists of even and odd numbers
print("Even numbers:", even_numbers)
print("Odd numbers:", odd_numbers)
```

```python
# Function to get a list of numbers smaller than a given value
def get_smaller_numbers(input_list, threshold):
    smaller_numbers = []  # List to store numbers smaller than the threshold

    # Iterate through each element in the input list
    for element in input_list:
        if element < threshold:
            smaller_numbers.append(element)  # Add the element to the list if it's smaller

    return smaller_numbers

# Given list of numbers and a threshold value
number_list = [8, 100, 20, 40, 3, 7]
threshold_value = 10

# Get the list of numbers smaller than the threshold using the function
smaller_numbers_list = get_smaller_numbers(number_list, threshold_value)

# Print the list of smaller numbers
print("Numbers smaller than", threshold_value, ":", smaller_numbers_list)
```

## 3. Largest Element in a List

Naive Soln.:

```python
# Function to find the maximum value in a list
def get_max_value(lst):
    for current_element in lst:        # Iterate through each element in the list
        for comparison_element in lst:  # Compare with each element in the list
            if comparison_element > current_element:  # Check if any element is greater
                break
        else:                           # If loop completes without break, no greater element found
            return current_element

    return None  # Return None if no maximum value is found

# Given list of integers
input_list = [10, 5, 20, 8, 15, 2, 30]

# Find and print the maximum value using the function
max_value = get_max_value(input_list)
print("Maximum value:", max_value)
```

## 4. Get Second Largest Element in List

Solution with Two Traversal

```python
def get_max(l):
    """
    Find the maximum element in the list.

    Args:
        l (list): The input list of integers.

    Returns:
        int or None: The maximum element in the list, or None if the list is empty.
    """
    if not l:
        return None
    else:
        res = l[0]                  # Assume l[0] is the maximum

        # Iterate through the list from index 1 to the last
        for i in range(1, len(l)):
            if l[i] > res:           # Check whether the current element is greater than res
                res = l[i]          # If the current element is greater, update res to current

        return res

def get_second_max(l):
    """
    Find the second maximum element in the list.

    Args:
        l (list): The input list of integers.

    Returns:
        int or None: The second maximum element in the list, or None if there are fewer than two elements.
    """
    if len(l) <= 1:
        return None

    largest = get_max(l)  # Get the largest element using the get_max function
    second_largest = None

    for x in l:
        if x != largest:
            if second_largest is None:
                second_largest = x
            else:
                second_largest = max(x, second_largest)

    return second_largest

# Given list of integers
input_list = [10, 5, 20, 8, 15, 2, 30]

# Find and print the second maximum value using the function
second_max_value = get_second_max(input_list)
print("Second maximum value:", second_max_value)
```

Solution with One Traversal

```python
def get_second_max(lst):
    """
    Find the second maximum element in the given list.

    Args:
        lst (list): The input list of integers.

    Returns:
        int or None: The second maximum element in the list, or None if there are fewer than two elements.
    """
    if len(lst) <= 1:
        return None

    largest = lst[0]  # Assume the first element is the largest
    second_largest = None

    # Iterate through the list starting from the second element
    for x in lst[1:]:
        if x > largest:          # Check if the current element is greater than the largest
            second_largest = largest  # Update the second largest with the previous largest
            largest = x         # Update the largest with the current element

        elif x != largest:      # If x is not the largest element
            if second_largest is None or second_largest < x:  # Check if x is greater than the current second largest
                second_largest = x  # Update the second largest with the current element

    return second_largest

# Given list of integers
input_list = [10, 5, 20, 8, 15, 2, 30]

# Find and print the second maximum value using the function
second_max_value = get_second_max(input_list)
print("Second maximum value:", second_max_value)
```

## 5. Check if List is Sorted (aesc)

```python
def is_sorted(lst):
    """
    Check if the given list is sorted in ascending order.

    Args:
        lst (list): The input list of elements.

    Returns:
        bool: True if the list is sorted, False otherwise.
    """
    sorted_lst = sorted(lst)  # Create a sorted version of the list

    if lst == sorted_lst:     # Compare the original list with the sorted list
        return True           # If they are the same, the list is sorted
    else:
        return False          # Otherwise, the list is not sorted

# Given list of elements
input_list = [10, 20, 30, 15, 40]

if is_sorted(input_list):
    print("Yes, the list is sorted.")
else:
    print("No, the list is not sorted.")
```

```python
# Original list
l = [10, 20, 30]

# Method 1: Using the reverse() method in-place
l.reverse()     # Reverse the given list in-place
print("Method 1 - Reversed list (in-place):", l)

# Method 2: Using the reversed() function to create a new reversed list
l = [10, 20, 30]   # Restore the original list
new_l = list(reversed(l))   # Create a new reversed list using the reversed() function
print("Method 2 - New reversed list (reversed()):", new_l)

# Method 3: Using slicing to create a new reversed list
l = [10, 20, 30]   # Restore the original list
new_l = l[::-1]             # Create a new reversed list using slicing
print("Method 3 - New reversed list (slicing):", new_l)

# Method 4: Custom Function
def reverse_array(arr, n):
    """
    Reverse the given array in-place.

    Args:
        arr (list): The input list of elements to be reversed.
        n (int): The size of the array.

    Returns:
        list: The reversed array.
    """
    start = 0      # Initialize the starting index
    end = n - 1    # Initialize the ending index

    while start < end:
        # Swap the elements at the start and end indices
        arr[start], arr[end] = arr[end], arr[start]
        start += 1   # Move the start index towards the end
        end -= 1     # Move the end index towards the start

    return arr      # Return the reversed array

# Given array
arr = [1, 2, 3, 4, 5]
n = len(arr)   # Calculate the size of the array

# Reverse the array using the function
reversed_arr = reverse_array(arr, n)
print("Method 4: Reversed array:", reversed_arr)
```

## 7. Removing Duplicates

Naive Approach

```python
def remove_duplicates(arr, n):
    """
    Remove duplicates from the given array and return the new length after removing duplicates.

    Args:
        arr (list): The input array containing duplicate elements.
        n (int): The original length of the array.

    Returns:
        int: The new length of the array after removing duplicates.
    """
    temp = [0] * n      # Create a temporary array to store non-duplicate elements
    temp[0] = arr[0]    # The first element is always non-duplicate
    res = 1             # Initialize the result (new length) as 1

    # Iterate through the array starting from the second element
    for i in range(1, n):
        if temp[res - 1] != arr[i]:  # Check if the current element is different from the last element in temp
            temp[res] = arr[i]       # Store the current element in temp
            res += 1                  # Increment the result (new length)

    # Copy the non-duplicate elements from temp back to arr
    for i in range(0, res):
        arr[i] = temp[i]

    return res  # Return the new length after removing duplicates

# Example usage
n = 7
arr = [10, 20, 20, 30, 30, 30, 30]
new_length = remove_duplicates(arr, n)
print("New length after removing duplicates:", new_length)
```

## 12. Mean And Median of Array


```python
def find_median(arr):
    """
    Calculate and return the median of the given array.

    Args:
        arr (list): The input list of numbers.

    Returns:
        int: The calculated median.
    """
    arr.sort()  # Sort the array
    length = len(arr)
    if length % 2 == 0:
        median = int((arr[length//2] + arr[(length//2)-1]) // 2)
    else:
        median = int(arr[length//2])
    return median

def find_mean(arr):
    """
    Calculate and return the mean of the given array.

    Args:
        arr (list): The input list of numbers.

    Returns:
        int: The calculated mean.
    """
    mean = int(sum(arr) / len(arr))
    return mean

# Given array
arr = [10, 5, 20, 8, 15, 2, 30]

# Calculate and print the median
median_value = find_median(arr)
print("Median:", median_value)

# Calculate and print the mean
mean_value = find_mean(arr)
print("Mean:", mean_value)
```

## 13. Count Smaller Than X



```python
def smaller_than_x(arr, n, x):
    """
    Count the number of elements in the array that are strictly smaller than x.

    Args:
        arr (list): The input list of non-negative integers.
        n (int): The size of the array.
        x (int): The threshold integer to compare with.

    Returns:
        int: The count of elements smaller than x.
    """
    # Create a list of elements that are smaller than x
    listed = [i for i in arr if i < x]

    # Return the count of elements in the list
    return len(listed)

# Given array and threshold integer
arr = [10, 5, 20, 8, 15, 2, 30]
x = 15

# Calculate and print the count of elements smaller than x
count = smaller_than_x(arr, len(arr), x)
print("Count of elements smaller than", x, ":", count)
```

## 14. Find Immediate Smaller Than X

```python
def immediate_smaller_than_x(arr, n, x):
    """
    Find the element in the array which is smaller than x and closest to it.

    Args:
        arr (list): The input list of positive integers.
        n (int): The size of the array.
        x (int): The threshold integer to compare with.

    Returns:
        int: The element in the array smaller than x and closest to it.
    """
    immediate_smaller = None

    # Iterate through the array to find the immediate smaller element
    for num in arr:
        if num < x:
            if immediate_smaller is None or (x - num) < (x - immediate_smaller):
                immediate_smaller = num

    return immediate_smaller

# Given array and threshold integer
arr = [10, 5, 20, 8, 15, 2, 30]
x = 16

# Find and print the element smaller than x and closest to it
result = immediate_smaller_than_x(arr, len(arr), x)
print("Immediate smaller element than", x, ":", result)
```

## 16. Find Immediate Greater Than X

```python
def immediate_greater(arr, n, x):
    """
    Find the value in the array that is greater than x and closest to it.

    Args:
        arr (list): The input list of positive integers.
        n (int): The size of the array.
        x (int): The threshold integer to compare with.

    Returns:
        int: The value in the array that is greater than x and closest to it, or -1 if no such value exists.
    """
    arr = sorted(arr)  # Sort the array in ascending order

    for i in arr:
        if i > x:
            return i  # Return the first value greater than x
            break
        else:
            continue

    # If no value greater than x is found, return -1
    if x > max(arr):
        return -1

# Given array and threshold integer
arr = [10, 5, 20, 8, 15, 2, 30]
x = 12

# Find and print the value greater than x and closest to it
result = immediate_greater(arr, len(arr), x)
print("Value greater than", x, "and closest to it:", result)
```

## 17. Find the Majority

```python
def find_most_frequent(arr, x, y):
    """
    Find the element that appears most frequently in the array. If both elements have the same frequency, return the smaller element.

    Args:
        arr (list): The input list of integers.
        x (int): The first element.
        y (int): The second element.

    Returns:
        int: The element that appears most frequently in the array. If both elements have the same frequency, return the smaller element.
    """
    freq_x = arr.count(x)  # Count the frequency of element x
    freq_y = arr.count(y)  # Count the frequency of element y

    if freq_x > freq_y:
        return x
    elif freq_y > freq_x:
        return y
    else:
        return min(x, y)

# Given array and elements x and y
arr = [1, 1, 2, 2, 3, 3, 4, 4, 4, 4, 5]
x = 4
y = 5

# Find and print the element that appears most frequently
result = find_most_frequent(arr, x, y)
print("Element that appears most frequently:", result)
```

## 18. Rotate Array

```python
def rotate_left(arr, n, d):
    """
    Rotate the given array to the left by D steps.

    Args:
        arr (list): The input list of elements to be rotated.
        n (int): The size of the array.
        d (int): The number of steps to rotate the array.

    Returns:
        list: The rotated array.
    """
    d = d % n  # Calculate the effective rotation steps within the array size

    # Slice the array into two parts and swap them
    rotated_arr = arr[d:] + arr[:d]

    return rotated_arr  # Return the rotated array

# Given array and rotation steps
arr = [1, 2, 3, 4, 5]
n = len(arr)  # Calculate the size of the array
d = 2        # Number of steps to rotate the array

# Rotate the array using the function
rotated_array = rotate_left(arr, n, d)
print("Rotated array:", rotated_array)
```
