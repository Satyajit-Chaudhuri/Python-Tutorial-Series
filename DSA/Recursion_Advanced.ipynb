# Rope Cutting Problem

image.png

image.png

```python
def maxpeices(n,a,b,c) :
    """
    Finds the maximum number of pieces a rope of length 'n' can be cut into,
    given three possible cut lengths 'a', 'b', and 'c'.

    Args:
        n: The length of the rope.
        a: The length of the first possible cut.
        b: The length of the second possible cut.
        c: The length of the third possible cut.

    Returns:
        The maximum number of pieces the rope can be cut into.
        Returns -1 if the rope cannot be cut into any pieces.
    """

    # Base case: If the rope length is 0, no cuts are possible.
    if n == 0 :
        return 0

    # Base case: If the rope length is negative, it indicates an invalid cut.
    if n <= -1 :
        return -1

    # Recursively explore the three possible cut lengths.
    # The maximum number of pieces is the maximum of the results obtained from
    # cutting the rope by lengths 'a', 'b', and 'c'.
    res = max(maxpeices(n-a,a,b,c),
              maxpeices(n-b,a,b,c),
              maxpeices(n-c,a,b,c))

    # If any of the recursive calls result in -1 (invalid cut), propagate -1.
    if res == -1 :
        return -1

    # Return the maximum number of pieces (including the current cut).
    return res + 1


n = 23
a = 11
b = 9
c = 12
print(maxpeices(n,a,b,c))
```

### After memoization

```python
def maxPieces(n, a, b, c, memo):
    if n == 0:
        return 0
    if n < 0:
        return -1
    if n in memo:
        return memo[n]

    res = max(maxPieces(n - a, a, b, c, memo),
              maxPieces(n - b, a, b, c, memo),
              maxPieces(n - c, a, b, c, memo))

    if res == -1:
        memo[n] = -1
    else:
        memo[n] = res + 1

    return memo[n]

def maxPiecesMemoized(n, a, b, c):
    memo = {}
    return maxPieces(n, a, b, c, memo)

# Test case
n = 23
a = 11
b = 9
c = 12
print(maxPiecesMemoized(n, a, b, c))  # Output: 2
```

# Recursive function to check if a string is palindrome

image.png

```python
# A recursive Python program
# to check whether a given
# number is palindrome or not

# A recursive function that
# check a str[s..e] is
# palindrome or not.
def isPalRec(st, s, e) :

	# If there is only one character
	if (s == e):
		return True

	# If first and last
	# characters do not match
	if (st[s] != st[e]) :
		return False

	# If there are more than
	# two characters, check if
	# middle substring is also
	# palindrome or not.
	if (s < e + 1) :
		return isPalRec(st, s + 1, e - 1);

	return True

def isPalindrome(st) :
	n = len(st)

	# An empty string is
	# considered as palindrome
	if (n == 0) :
		return True

	return isPalRec(st, 0, n - 1);


# Driver Code
st = "geeg"
if (isPalindrome(st)) :
	print("Yes")
else :
	print ("No")
```

Time Complexity: O(n)

Auxiliary Space: O(n)

```python
def THO(n, A, B, C):
    if n == 1:
        print("Move 1 from", A, "to", C)
    else:
        THO(n - 1, A, C, B)
        print("Move", n, "from", A, "to", C)
        THO(n - 1, B,A, C)


THO(3, 'A', 'B','C' )
```

# Josephus Problem in Python

image.png

image.png

```python
def jos(n, k):
    # Base case: If there is only one person left, they are the survivor (position 0)
    if n == 1:
        return 0
    else:
        # Recursive case: Find the position of the survivor in the smaller circle
        # and adjust the position according to the current circle size
        return (jos(n - 1, k) + k) % n

# Call the josephus function with 5 people and every 3rd person is eliminated
print(jos(5, 3))  # Output the position of the survivor
```

Time Complexity: O(N)

Auxiliary Space: O(N) the space used in recursion call stack

# Subset Sum Problem

image.png

image.png

image.png

```python
def countSubsetSum(arr,n,sum):
    if n==0:
        if sum==0:
            return 1
        else:
            return 0

    return countSubsetSum(arr,n-1,sum) + countSubsetSum(arr,n-1,sum-arr[n-1])

arr = [1,2,3,5,6,7]
n=6
sum = 8

print(countSubsetSum(arr,n,sum))
```

**Complexity Analysis:** The above solution may try all subsets of given set in worst case. Therefore time complexity of the above solution is exponential. The problem is in-fact NP-Complete (There is no known polynomial time solution for this problem).

# Printing all Permutations

```python
# Python program to print all permutations with
# duplicates allowed

def toString(List):
	return ''.join(List)

# Function to print permutations of string
# This function takes three parameters:
# 1. String
# 2. Starting index of the string
# 3. Ending index of the string.
def permute(a, l, r):
	if l==r:
		print (toString(a))
	else:
		for i in range(l,r):
			a[l], a[i] = a[i], a[l]
			permute(a, l+1, r)
			a[l], a[i] = a[i], a[l] # backtrack

# Driver program to test the above function
string = "ABC"
n = len(string)
a = list(string)
permute(a, 0, n)
```

# Print all subsequences of a string

```python
# Below is the implementation of the above approach
def printSubsequence(input, output):

	# Base Case
	# if the input is empty print the output string
	if len(input) == 0:
		print(output, end=' ')
		return

	# output is passed with including the
	# 1st character of input string
	printSubsequence(input[1:], output+input[0])

	# output is passed without including the
	# 1st character of input string
	printSubsequence(input[1:], output)


# Driver code
# output is set to null before passing in
# as a parameter
output = ""
input = "abcd"

printSubsequence(input, output)
```
